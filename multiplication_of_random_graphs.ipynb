```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
n = 10 # number of nodes
p = 0.1 # number of edges

adjacency_matrixA = np.zeros([n, n])
adjacency_matrixA[np.random.random([n,n]) < p] = 1
```

```python
G = nx.from_numpy_matrix(adjacency_matrixA, create_using=nx.MultiDiGraph())

fig=plt.figure(figsize=(18, 10), dpi= 80, facecolor='w', edgecolor='k')

plt.subplot(1,2,1)
nx.draw_circular(G)
plt.axis('equal')

plt.subplot(1,2,2)
plt.imshow(adjacency_matrixA)
```
